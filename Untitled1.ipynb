```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data= pd.read_excel('data.xlsx').values
x = data[:,:-1]
y = data[:,-1]
x,y
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
y_pred = model.predict(x)
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred, color='red')
model.intercept_
model.coef_
```
